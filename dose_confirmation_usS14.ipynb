```python
# Task 14
# Dose confirmation notification

import state


def notifications_allowed_for_confirmation():
    return SESSION.get("notifications_on", True)


def record_dose_as_taken(user_email, medicine_id, medicine_name, taken_time_text):

# Task 14
# store that a dose was taken



    if user_email is None:
        return

    if user_email not in DOSE_HISTORY:
        DOSE_HISTORY[user_email] = []

    entry = {
        "medicine_id": medicine_id,
        "medicine_name": medicine_name,
        "taken_at": str(taken_time_text).strip()
    }

    DOSE_HISTORY[user_email].append(entry)


def get_confirmation_status_for_medicine(user_email, medicine_id):

   # Task 14
   # let the frontend check if there is at least one recorded taken dose for this medicine in the current session or user


    if user_email is None:
        return False

    history = DOSE_HISTORY.get(user_email, [])

    for entry in history:
        if entry["medicine_id"] == medicine_id:
            return True

    return False


def show_dose_confirmation_notification(user_email, medicine_id, medicine_name):

   # Task 14
   # after a dose is marked as taken, call this to
    # 1 record it in history
    # 2 show a confirmation notification in the app (if notifications are on)


    record_dose_as_taken(
        user_email=user_email,
        medicine_id=medicine_id,
        medicine_name=medicine_name,
        taken_time_text="now"
    )

    if not notifications_allowed_for_confirmation():
        return

    print("Dose confirmation for user:", user_email)
    print("Medicine ID:", medicine_id)
    print("Medicine name:", medicine_name)
    print("Status: marked as taken")
    print(" -> would show a confirmation notification in the app UI")
```
