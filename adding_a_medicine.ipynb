```python
from datetime import datetime

class Medicine:
    def __init__(self, name, expiry_date, doses_per_day, dose_size, medicine_type, additional_info, scheduled_time=None):
        self.name = name
        # Convert expiry_date string to datetime object
        if isinstance(expiry_date, str):
            self.expiry_date = datetime.strptime(expiry_date, '%Y-%m-%d')
        else:
            self.expiry_date = expiry_date
        self.doses_per_day = doses_per_day
        self.dose_size = dose_size
        self.medicine_type = medicine_type
        self.additional_info = additional_info
        self.scheduled_time = scheduled_time # Added scheduled_time attribute

    def display_info(self):
        print(f"Medicine Name: {self.name}")
        # Format datetime object for display
        print(f"Expiry Date: {self.expiry_date.strftime('%Y-%m-%d')}")
        print(f"Doses Per Day: {self.doses_per_day}")
        print(f"Dose Size: {self.dose_size}")
        print(f"Medicine Type: {self.medicine_type}")
        print(f"Additional Information: {self.additional_info}")
        if self.scheduled_time:
            print(f"Scheduled Time: {self.scheduled_time.strftime('%H:%M')}")
        else:
            print("Scheduled Time: Not set")
```

```python
def edit_medicine_details(medicine):
    print(f"\n--- Editing Medicine: {medicine.name} ---")

    # Edit Name
    new_name = input(f"Enter new medicine name (current: {medicine.name}, leave blank to keep): ")
    if new_name:
        medicine.name = new_name

    # Edit Expiry Date
    while True:
        new_expiry_date_str = input(f"Enter new expiry date (YYYY-MM-DD, current: {medicine.expiry_date.strftime('%Y-%m-%d')}, leave blank to keep): ")
        if not new_expiry_date_str:
            break
        try:
            medicine.expiry_date = datetime.strptime(new_expiry_date_str, '%Y-%m-%d')
            break
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

    # Edit Doses Per Day
    while True:
        new_doses_per_day_str = input(f"Enter new number of doses per day (current: {medicine.doses_per_day}, leave blank to keep): ")
        if not new_doses_per_day_str:
            break
        try:
            new_doses = int(new_doses_per_day_str)
            if new_doses > 10:
                print("Number of doses per day cannot be over 10. Please enter a valid number.")
            else:
                medicine.doses_per_day = new_doses
                break
        except ValueError:
            print("Invalid input. Please enter a valid number for doses per day.")

    # Edit Dose Size
    new_dose_size = input(f"Enter new dose size (e.g., 500mg, 5ml, current: {medicine.dose_size}, leave blank to keep): ")
    if new_dose_size:
        medicine.dose_size = new_dose_size

    # Edit Medicine Type
    new_medicine_type = input(f"Enter new medicine type (e.g., Tablet, Syrup, Capsule, current: {medicine.medicine_type}, leave blank to keep): ")
    if new_medicine_type:
        medicine.medicine_type = new_medicine_type

    # Edit Additional Info
    new_additional_info = input(f"Enter new information key (e.g., Storage, Instructions, current: {medicine.additional_info}, leave blank to keep): ")
    if new_additional_info:
        medicine.additional_info = new_additional_info

    print("\n--- Medicine Details Updated ---")
    medicine.display_info()
```

```python
def get_medicine_details_from_user():
    # Get medicine details from user input
    name = input("Enter medicine name: ")

    while True:
        expiry_date_str = input("Enter expiry date (YYYY-MM-DD): ")
        try:
            expiry_date = datetime.strptime(expiry_date_str, '%Y-%m-%d')
            break
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

    while True:
        try:
            doses_per_day = int(input("Enter number of doses per day: "))
            if doses_per_day > 10:
                print("Number of doses per day cannot be over 10. Please enter a valid number.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a valid number for doses per day.")

    dose_size = input("Enter dose size (e.g., 500mg, 5ml): ")
    medicine_type = input("Enter medicine type (e.g., Tablet, Syrup, Capsule): ")
    additional_info = input("Enter information key (e.g., Storage, Instructions) or leave blank: ")

    user_medicine = Medicine(
        name=name,
        expiry_date=expiry_date,
        doses_per_day=doses_per_day,
        dose_size=dose_size,
        medicine_type=medicine_type,
        additional_info=additional_info
    )
    return user_medicine
```

```python
def main_menu():
    #medicines = load_medicines()

    while True:
        print("--- Medicine Management System ---")
        print("1. Add Medicine")
        print("2. Edit Medicine")
        print("3. Display All Medicines")
        print("4. Check Expired Medicines")
        print("5. Delete Medicine")
        print("6. Link Time to Medicine")
        print("7. Check Scheduled Medicines")
        print("8. Save Medicines")
        print("9. Load Medicines")
        print("10. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            print("You chose to add a medicine.")
            new_medicine = get_medicine_details_from_user()
            medicines.append(new_medicine)
            print(f"Medicine '{new_medicine.name}' added successfully!")
        elif choice == '2':
            print("You chose to edit a medicine.")
            if not medicines:
                print("No medicines to edit. Please add a medicine first.")
                continue

            print("--- Available Medicines ---")
            for i, med in enumerate(medicines):
                print(f"{i + 1}. {med.name}")

            while True:
                try:
                    edit_choice = int(input("Enter the number of the medicine you wish to edit: "))
                    if 1 <= edit_choice <= len(medicines):
                        selected_medicine = medicines[edit_choice - 1]
                        edit_medicine_details(selected_medicine)
                        break
                    else:
                        print("Invalid number. Please enter a valid number from the list.")
                except ValueError:
                    print("Invalid input. Please enter a number.")

        elif choice == '3':
            print("You chose to display all medicines.")
            if not medicines:
                print("No medicines to display. Please add a medicine first.")
                continue
            print("--- All Medicines ---")
            for med in medicines:
                med.display_info()
                print("---------------------")
        elif choice == '4':
            print("--- Checking for Expired Medicines ---")
            if not medicines:
                print("No medicines to check. Please add a medicine first.")
                continue

            expired_found = False
            for med in medicines:
                if is_medicine_expired(med):
                    print(f"WARNING: {med.name} is expired (Expiry: {med.expiry_date.strftime('%Y-%m-%d')})")
                    expired_found = True
            if not expired_found:
                print("No expired medicines found.")
        elif choice == '5':
            print("You chose to delete a medicine.")
            delete_medicine(medicines)
        elif choice == '6':
            print("You chose to link a time to a medicine.")
            if not medicines:
                print("No medicines available to link a time to. Please add a medicine first.")
                continue

            print("--- Available Medicines ---")
            for i, med in enumerate(medicines):
                print(f"{i + 1}. {med.name}")

            while True:
                try:
                    link_choice = int(input("Enter the number of the medicine you wish to link a time to: "))
                    if 1 <= link_choice <= len(medicines):
                        selected_medicine = medicines[link_choice - 1]
                        user_time = get_time_from_user()
                        selected_medicine.scheduled_time = user_time
                        print(f"Scheduled time {user_time.strftime('%H:%M')} linked to '{selected_medicine.name}' successfully!")
                        break
                    else:
                        print("Invalid number. Please enter a valid number from the list.")
                except ValueError:
                    print("Invalid input. Please enter a number.")
        elif choice == '7':
            check_scheduled_medicines(medicines)
        elif choice == '8': # New Save option
            save_medicines(medicines)
        elif choice == '9': # New Load option
            loaded_data = load_medicines()
            if loaded_data:
                medicines = loaded_data
                print("Medicine data loaded successfully!")
            else:
                print("No data to load or an error occurred during loading.")
        elif choice == '10': # Shifted Exit option
            print("Exiting Medicine Management System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
from datetime import datetime

def is_medicine_expired(medicine):

    today = datetime.now().date()
    if medicine.expiry_date.date() < today:
        return True
    else:
        return False


```

```python
import smtplib
from email.message import EmailMessage

def send_expiration_email(medicine, recipient_email):

    sender_email = ""  # TODO: Replace with your email
    sender_password = ""

    msg = EmailMessage()
    msg['Subject'] = f"Urgent: Medicine Expiration Alert for {medicine.name}"
    msg['From'] = sender_email
    msg['To'] = recipient_email

    msg.set_content(f"""
Dear User,

This is an urgent notification that your medicine '{medicine.name}' has expired.

Medicine Details:
Name: {medicine.name}
Expiry Date: {medicine.expiry_date.strftime('%Y-%m-%d')}
Doses Per Day: {medicine.doses_per_day}
Dose Size: {medicine.dose_size}
Medicine Type: {medicine.medicine_type}
Additional Info: {medicine.additional_info}

Please take appropriate action regarding this expired medication.

Sincerely,
Your Medicine Management System
""")

    try:
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(sender_email, sender_password)
            smtp.send_message(msg)
        print(f"Expiration email for {medicine.name} sent successfully to {recipient_email}!")
    except Exception as e:
        print(f"Failed to send expiration email for {medicine.name}. Error: {e}")

```

```python
def delete_medicine(medicines):
    print("\n--- Delete Medicine ---")
    if not medicines:
        print("No medicines to delete.")
        return

    while True:
        print("\n--- Available Medicines ---")
        for i, med in enumerate(medicines):
            print(f"{i + 1}. {med.name}")
        print(f"{len(medicines) + 1}. Back to Main Menu")

        try:
            delete_choice = int(input("Enter the number of the medicine you wish to delete, or go back: "))
            if 1 <= delete_choice <= len(medicines):
                removed_medicine = medicines.pop(delete_choice - 1)
                print(f"Medicine '{removed_medicine.name}' deleted successfully!")
                break
            elif delete_choice == len(medicines) + 1:
                print("Returning to main menu.")
                break
            else:
                print("Invalid number. Please enter a valid number from the list.")
        except ValueError:
            print("Invalid input. Please enter a number.")
```

```python
from datetime import datetime

def get_time_from_user():
    while True:
        time_str = input("Enter a time (HH:MM): ")
        try:
            # Parse the time string into a datetime object, then extract the time part
            user_time = datetime.strptime(time_str, '%H:%M').time()
            return user_time # Re-added this line to return the time
        except ValueError:
            print("Invalid time format. Please enter the time in HH:MM format.")
```

```python
from datetime import datetime

def check_scheduled_medicines(medicines):
    print("\n--- Checking Scheduled Medicines ---")
    if not medicines:
        print("No medicines added yet.")
        return

    current_time = datetime.now().time()
    found_scheduled_medicine_for_now = False

    for med in medicines:
        if med.scheduled_time:
            if med.scheduled_time.hour == current_time.hour and med.scheduled_time.minute == current_time.minute:
                print(f"ALERT: It's time for '{med.name}'! Scheduled for {med.scheduled_time.strftime('%H:%M')}.")
                med.display_info()
                found_scheduled_medicine_for_now = True

    if not found_scheduled_medicine_for_now:
        print("No scheduled medicines found for the current time.")
```

```python
import json
from datetime import datetime, time

def save_medicines(medicines, filename="medicines_data.json"):
    serializable_medicines = []
    for med in medicines:
        med_dict = {
            "name": med.name,
            "expiry_date": med.expiry_date.strftime('%Y-%m-%d'),
            "doses_per_day": med.doses_per_day,
            "dose_size": med.dose_size,
            "medicine_type": med.medicine_type,
            "additional_info": med.additional_info,
            "scheduled_time": med.scheduled_time.strftime('%H:%M') if med.scheduled_time else None
        }
        serializable_medicines.append(med_dict)

    try:
        with open(filename, 'w') as f:
            json.dump(serializable_medicines, f, indent=4)
        print(f"Medicine data saved to {filename}")
    except Exception as e:
        print(f"Error saving medicine data: {e}")

```

```python

def load_medicines(filename="medicines_data.json"):
    """Loads medicine objects from a JSON file."""
    try:
        with open(filename, 'r') as f:
            serializable_medicines = json.load(f)

        loaded_medicines = []
        for med_dict in serializable_medicines:

            expiry_date_dt = datetime.strptime(med_dict["expiry_date"], '%Y-%m-%d')

            scheduled_time_obj = None
            if med_dict["scheduled_time"]:
                scheduled_time_obj = datetime.strptime(med_dict["scheduled_time"], '%H:%M').time()

            medicine = Medicine(
                name=med_dict["name"],
                expiry_date=expiry_date_dt,
                doses_per_day=med_dict["doses_per_day"],
                dose_size=med_dict["dose_size"],
                medicine_type=med_dict["medicine_type"],
                additional_info=med_dict["additional_info"],
                scheduled_time=scheduled_time_obj
            )
            loaded_medicines.append(medicine)
        print(f"Medicine data loaded from {filename}")
        return loaded_medicines
    except FileNotFoundError:
        print(f"No medicine data file found at {filename}. Starting with an empty list.")
        return []
```
