```python
# Task 12
# notifications and reminder times- 12.1 and 12.2

import state

email = state.get_logged_in_email()
state.SESSION[ "logged_in"] = True
state.SESSION["email"] = "test@example.com"

SESSION = {
    "logged_in": False,
    "email": None,
    "current_screen": "login",
    "notifications_on": True,              # single on or off for all app notifications
    "default_notification_time": "08:00"
}

# 12.1 global default notification time and sending

def set_global_notification_time(from_html_time_value):
    if from_html_time_value is None:
        return

    time_text = str(from_html_time_value).strip()
    if time_text == "":
        return

    SESSION["default_notification_time"] = time_text


def get_global_notification_time_for_html():
    return SESSION.get("default_notification_time", "08:00")


def get_global_notification_plan():
    plan = {
        "time": SESSION.get("default_notification_time", "08:00"),
        "notifications_on": SESSION.get("notifications_on", True)
    }
    return plan


def send_global_notification(message_text):
    plan = get_global_notification_plan()

    time_to_send = plan["time"]
    allowed = plan["notifications_on"]

    if not allowed:
        return

    print("Global notification time:", time_to_send)
    print("Message:", message_text)
    print(" -> would show this notification in the app UI")







# 12.2 custom reminders per medication

MEDICATION_REMINDERS = {
    # "user_email": [
    #  something like {"medicine_id": "med1", "name": "Paracetamol", "custom_time": "09:30"},
    # ]
}


def set_custom_reminder_for_medicine(user_email, medicine_id, medicine_name,
                                     custom_time_text):
    if user_email is None:
        return

    if user_email not in MEDICATION_REMINDERS:
        MEDICATION_REMINDERS[user_email] = []

    if custom_time_text is not None:
        custom_time_text = str(custom_time_text).strip()
        if custom_time_text == "":
            custom_time_text = None
    else:
        custom_time_text = None

    found = False
    for entry in MEDICATION_REMINDERS[user_email]:
        if entry["medicine_id"] == medicine_id:
            entry["name"] = medicine_name
            entry["custom_time"] = custom_time_text
            found = True
            break

    if not found:
        MEDICATION_REMINDERS[user_email].append({
            "medicine_id": medicine_id,
            "name": medicine_name,
            "custom_time": custom_time_text
        })


def get_effective_time_for_medicine(user_email, medicine_id):
    global_time = SESSION.get("default_notification_time", "08:00")

    if user_email is None:
        return global_time

    reminders = MEDICATION_REMINDERS.get(user_email, [])

    for entry in reminders:
        if entry["medicine_id"] == medicine_id:
            if entry["custom_time"] is not None:
                return entry["custom_time"]
            break

    return global_time


def notifications_allowed_for_medicine():
    return SESSION.get("notifications_on", True)


def send_custom_medicine_notification(user_email, medicine_id, message_text):
    if not notifications_allowed_for_medicine():
        return

    time_to_send = get_effective_time_for_medicine(user_email, medicine_id)

    print("Medicine notification for user:", user_email)
    print("Medicine ID:", medicine_id)
    print("Send at:", time_to_send)
    print("Message:", message_text)
    print(" -> would show this notification in the app UI")
```
